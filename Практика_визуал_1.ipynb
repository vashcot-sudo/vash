# Задание

* Создайте датасет, который содержит 7 переменных: 3 категориальных, 3 числовых, 1 дата/время. Тему выберите сами.

* Постройте гипотезу.

* Проверьте гипотезу с помощью визуализации.

* Сделайте выводы.

```python
import pandas as pd
import numpy as np

np.random.seed(0)
# Создание категорийных признаков
# Создание категорийной переменной с тремя категориями Смартфон, Ноутбук, Наушники
gadgets = np.random.choice(['Смартфон', 'Ноутбук', 'Наушники'], size=40, replace=True)
# Создание категорийной переменной с двумя категориями, повторяющимися по очереди: голубой, зеленый
colour = ['голубой',"зеленый"]*20
brand = ["Apple", 'Samsung', 'Lenovo', 'Honor', 'Asus'] * 8

# Создание числовых признаков
number = np.random.randint(7,25,size=40) # рандомно 40 целых чисел от 7 до 25
quantity_stock = np.random.randint(0, 100, size=40) # количество на складе (рандомно 40 целых цисел от 0 до 100)
#price = np.random.normal(85000,93000,size=40)
price = np.round(np.random.uniform(85000,93000,40), 2)
#price=[85000,86000,87000,91000,93000,88000,89000,90000,91000,92000,93000,85000,86000,87000,90000,92000,88000,89000,90000,91000,92000,93000,91000,87000,120000,90000,88000,93000,87000,85000,92000,90000,91000,85000,87000,92000,87000,120000,87000,91000,]
numerical = np.random.randn(40) # 40 рандомных чисел

# Создание признака дата/время
priv = pd.to_datetime(pd.date_range(start='2025-07-19', periods=40, freq='D'))

# Создание датасета
data = {
    'gadgets': gadgets,
    'colour': colour,
    'brand': brand,
    'number': number,
    'quantity_stock': quantity_stock,
    'price': price,
    'numerical': numerical,
    'priv':priv
}

df = pd.DataFrame(data)

# вывод датасета на экран
display(df)
```

Гипотеза: количество продаж не зависит от бренда и цвета гаджета.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.violinplot(x='brand', y='number', data=df, palette='Set3', hue=colour)
plt.title('Зависимость количества продаж от бренда и цвета гаджета')
plt.xlabel('Бренд')
plt.ylabel('Количество продаж')
plt.tight_layout()
plt.show()
```

Гипотеза не подтверждается. График показывает, что количество продаж зависит и от бренда, и от цвета гаджета
